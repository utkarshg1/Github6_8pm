```python
import yfinance as yf
ticker = 'SBIN.NS'
share_df = yf.download(tickers=ticker, period='5y', interval='1mo')
share_df.head()
```

```python
share_df.index
```

```python
type(share_df.index)
```

```python
df_close = share_df[['Close']]
```

```python
df_close
```

### Get the Time Series decomposition and forecast for above data

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(df_close['Close'])
plt.xlabel('Month')
plt.ylabel('Closing Price')
plt.title('SBI Stock Price')
plt.grid()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
Q = seasonal_decompose(df_close['Close'])
```

```python
Q.plot()
```

```python
Q.observed.plot()
Q.trend.plot()
```

```python
Q.seasonal.plot()
```

```python
Q.resid.plot()
```
